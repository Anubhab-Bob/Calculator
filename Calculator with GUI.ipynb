```python
'''



Step by Step Description 


1. User clicks a number button
    Note 1: User can input single as well as multiple digits as number

2. User clicks a math button
    Note 1: Make sure the 1st number button is clicked beforehand
            Otherwise show error
    Note 2: If multiple math buttons are clicked accept only the last one
    Note 3: Handle Exceptions
    
3. User clicks another number button
    Note 1: User can input single as well as multiple digits as number
    
4. User clicks '=' and the result is shown on the screen
    Note 1: Make sure a math button has been clicked beforehand
            Otherwise show error
    Note 2: Check which button was clicked and act accordingly


'''
```

```python
# Importing required files
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
import operator
import math
```

```python
class Calculator :
    
    
    calc_value = 0.0             # Stores the current value to display in the entry
    
    # Will define if this was the last math button clicked
    add = False
    diff = False
    mult = False
    div = False
    mod = False
    sq = False
    cb = False
    exp = False
    op = False
    
    
    def __init__(self, root) :
        
        # Will hold the changing value stored in the entry
        self.entry = StringVar(root, value="")
        
        root.title("Calculator")               # Application Name

        root.geometry("554x370+300+300")       # Sizing the application9

        root.resizable(width=False, height=False)    # Block resizing of Window
        
        frame = Frame(root)                    # Adding frame to support widgets
        
        # Customize the styling for the buttons and entry
        style = ttk.Style()
        style.configure("TButton",
                        foreground="midnight blue",
                        font="Times 15",
                        padding=5)
        
        style.configure("TEntry",
                        foreground="DarkOrchid4",
                        font="Serif 18",
                        padding=10)
        
        # Create the text entry box
        self.entry = ttk.Entry(root, 
                               textvariable=self.entry,
                               width=50 )
        self.entry.grid(row=0,  columnspan=4, padx=5, pady=15)
        
        
         # ----- 1st Row -----
 
        self.button7 = ttk.Button(root, text="7", command=lambda: self.button_press('7')).grid(row=1, column=0, padx=5, pady=5) 
        self.button8 = ttk.Button(root, text="8", command=lambda: self.button_press('8')).grid(row=1, column=1, padx=5, pady=5) 
        self.button9 = ttk.Button(root, text="9", command=lambda: self.button_press('9')).grid(row=1, column=2, padx=5, pady=5) 
        self.button_div = ttk.Button(root, text="/", command=lambda: self.math_button_press('/')).grid(row=1, column=3, padx=5, pady=5)
        
        # ----- 2nd Row -----
 
        self.button4 = ttk.Button(root, text="4", command=lambda: self.button_press('4')).grid(row=2, column=0, padx=5, pady=5) 
        self.button5 = ttk.Button(root, text="5", command=lambda: self.button_press('5')).grid(row=2, column=1, padx=5, pady=5) 
        self.button6 = ttk.Button(root, text="6", command=lambda: self.button_press('6')).grid(row=2, column=2, padx=5, pady=5) 
        self.button_mul = ttk.Button(root, text="*", command=lambda: self.math_button_press('*')).grid(row=2, column=3, padx=5, pady=5)
        
        # ----- 3rd Row -----
 
        self.button1 = ttk.Button(root, text="1", command=lambda: self.button_press('1')).grid(row=3, column=0, padx=5, pady=5) 
        self.button2 = ttk.Button(root, text="2", command=lambda: self.button_press('2')).grid(row=3, column=1, padx=5, pady=5) 
        self.button3 = ttk.Button(root, text="3", command=lambda: self.button_press('3')).grid(row=3, column=2, padx=5, pady=5) 
        self.button_add = ttk.Button(root, text="+", command=lambda: self.math_button_press('+')).grid(row=3, column=3, padx=5, pady=5)
        
        # ----- 4th Row -----
 
        self.button_mod = ttk.Button(root, text="%", command=lambda: self.math_button_press('%')).grid(row=4, column=0, padx=5, pady=5) 
        self.button0 = ttk.Button(root, text="0", command=lambda: self.button_press('0')).grid(row=4, column=1, padx=5, pady=5) 
        self.button_equal = ttk.Button(root, text="=", command=lambda: self.equal_button_press()).grid(row=4, column=2, padx=5, pady=5)
        self.button_diff = ttk.Button(root, text="-", command=lambda: self.math_button_press('-')).grid(row=4, column=3, padx=5, pady=5)
        
        # ----- 5th Row -----
 
        self.button_clear = ttk.Button(root, text="AC", command=lambda: self.button_press('AC')).grid(row=5, column=0, padx=5, pady=5) 
        self.button_sqRt = ttk.Button(root, text=u"\u221A", command=lambda: self.math_button_press('sq')).grid(row=5, column=1, padx=5, pady=5) 
        self.button_cbRt = ttk.Button(root, text=u"\u221B", command=lambda: self.math_button_press('cb')).grid(row=5, column=2, padx=5, pady=5) 
        self.button_exp = ttk.Button(root, text="^", command=lambda: self.math_button_press('exp')).grid(row=5, column=3, padx=5, pady=5) 

        # ----- 6th Row -----
 
        self.button_dot = ttk.Button(root, text=".", command=lambda: self.button_press('.')).grid(row=6, column=1, padx=5, pady=5) 
        self.button_sign = ttk.Button(root, text="(-)", command=lambda: self.math_button_press('sign')).grid(row=6, column=2, padx=5, pady=5) 
        
        
        
    # Returns True or False if the string is a number    
    def is_number(self, value) :
        
        try :
            x = float(value)
            return True           # Return true if there is a number, otherwise generates value error
        
        except ValueError :
            return False          # Return false if input is not number
        
        
        
    def button_press(self, button) :
        
        if button == 'AC' :
            self.entry.delete(0, "")               # Clear the entry box
            self.op = False
            return
        
        #elif self.is_number(str(self.entry.get())) and button == '-' :
        if self.op == True and self.is_number(str(self.entry.get())) == False :
            self.entry.delete(0, "end")               # Clear the entry box
        
        entry_val = self.entry.get()           # Get the current value in the entry
        
        # Put the new value to the right of it
        # If it was 1 and 2 is pressed it is now 12
        # Otherwise the new number goes on the leftmost position
        entry_val += button
        
        self.entry.delete(0, "")               # Clear the entry box
        
        self.entry.insert(0, entry_val)        # Insert the new value going from left to right
        

        
    def math_button_press(self, button) :
        
        
        # Only do anything if entry currently contains a number or operator
        if self.is_number(str(self.entry.get())) or self.op == True :
            
            self.op = True
            # make false to cancel out previous math button click
            self.add = False
            self.diff = False
            self.mult = False
            self.div = False
            self.mod = False
            self.sq = False
            self.cb = False
            self.exp = False
            

            
            # Get the value out of the entry box for the calculation
            if self.is_number(str(self.entry.get())) :
                self.calc_value = float(self.entry.get())
            
                
            self.entry.delete(0, "end")               # Clear the entry box
            
            if button == '+' :
                self.add = True
                self.entry.insert(0, "+")
            elif button == '-' :
                self.diff = True
                self.entry.insert(0, "-")
            elif button == '/' :
                self.div = True
                self.entry.insert(0, "/")
            elif button == '*' :
                self.mult = True
                self.entry.insert(0, "*")
            elif button == '%' :
                self.mod = True
                self.entry.insert(0, "%")
            elif button == 'exp' :
                self.exp = True
            elif button == 'sq' :
                self.entry.insert(0, math.sqrt(self.calc_value))
                self.op = False
            elif button == 'cb' :
                self.entry.insert(0, (int(self.calc_value) ** (1/3)))
                self.op = False
            elif button == 'sign' :
                self.calc_value = -(self.calc_value)
                self.op = False
                self.entry.insert(0, self.calc_value)
            
        
    # Performs a mathematical operation by taking the value before
    # the math button is clicked and the current value. Then perform
    # the right calculation by checking what math button was clicked
    # last
    def equal_button_press(self) :
 
        # Make sure a math button was clicked
        solution = 0
        if self.op :
 
            if self.add == True :
                solution = self.calc_value + float(self.entry.get())
            elif self.diff :
                solution = self.calc_value - float(self.entry.get())
            elif self.mult :
                solution = self.calc_value * float(self.entry.get())
            elif self.div :
                solution = self.calc_value / float(self.entry.get())
            elif self.mod :
                solution = operator.mod(self.calc_value, float(self.entry.get()))
            elif self.exp :
                solution = self.calc_value ** float(self.entry.get())
            self.op = False
        self.entry.delete(0, 'end')            # 'end' makes deletion till the end of the value
        self.entry.insert(0, "{:.15g}".format(solution))       # Print output upto 15 significant digits
        
        
root = Tk()

calc = Calculator(root)
root.mainloop()
```
